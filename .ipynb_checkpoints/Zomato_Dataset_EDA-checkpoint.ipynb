# Zomato Dataset Exploratory Data Analysis

In this task we will performing an exploratory data analysis on the Zomato Dataset.<br>


```python
# Importing all the modules required for analysing the data.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading the csv file containing the data using pandas
df = pd.read_csv('Zomatodataset/zomato.csv', encoding = 'latin-1')
df.head()
```

```python
# Columns in the dataset
df.columns
```

```python
# Information about the dataset columns such data type, count, etc
df.info()
```

```python
# Description of the dataset
df.describe()
```

```python
df.shape
```

The dataset contains 9551 rows and 21 columns in total.

```python
# Checking for null values
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum() > 0]
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
# Reading the Country-Code excel file from the Zomato dataset
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
# All the columns of the csv file and the excel file
df.columns, df_country.columns
```

```python
# Merging the both csv and excel files into one and renaming as final_df
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
```

```python
final_df.head()
```

```python
# Checking data types
final_df.dtypes
```

```python
final_df.columns
```

```python
# Checking for the countries in which Zomato is available
final_df.Country.value_counts().index
```

```python
country_names = final_df.Country.value_counts().index
```

```python
# Checking for the number of Zomato restaurants in the countries
final_df.Country.value_counts().values
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# Pie Chart - Top 3 countries that use Zomato
plt.pie(country_values[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

#### Observation:
Zomato maximum records or transaction are from India followed by United States and United Kingdom.

```python
# Ratings given by the people distinguished according to the rating colour and rating text
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0 : 'Rating Count'})
ratings
```

## Conclusion on ratings
When the rating is between 4.5 to 4.9 --> Excellent<br>
When the ratings are between 4.0 to 4.5 --> Very Good<br>
When the ratings are between 3.5 to 3.9 --> Good<br>
When the ratings are between 2.5 to 3.4 --> Average<br>
When the ratings are between 1.8 to 2.4 --> Poor<br>
When the ratings are 0 --> Not rated<br>

```python
# Barplot of Ratings according to the rating colour
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = ratings, palette = ['pink', 'red', 'orange', 'yellow', 'green', 'green'])
```

### Observation:
1. Note rated count is very high
2. Maximum number of rating are between 2.5 and 3.4

```python
# Count plot ratings and rating colour
sns.countplot(x = "Rating color", data = ratings, hue = 'Rating color', palette = ['pink', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
final_df.groupby(['Aggregate rating', 'Country']).size().reset_index()
```

### Which currency is used by which country ?

```python
final_df[['Country', 'Currency']].groupby(['Country', 'Currency']).size().reset_index()
```

### Which countries do have online deliveries option?

```python
final_df[['Has Online delivery', 'Country']].groupby(['Has Online delivery','Country']).size().reset_index()
```

### A Pie Chart for cities distribution

```python
city_values = final_df.City.value_counts().values
city_labels = final_df.City.value_counts().index
```

```python
plt.pie(city_values[:5], labels = city_labels[:5],autopct = '%1.2f%%')
```

### Top 10 cuisines

```python
final_df[['Cuisines', 'Aggregate rating']].sort_values('Aggregate rating', ascending = False).groupby(['Cuisines', 'Aggregate rating'], sort = False).size().reset_index().head(10)
```

### Plotting the locations of Zomato restaurants on World Map

```python
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

geometry = [Point(xy) for xy in zip(final_df['Longitude'], final_df['Latitude'])]
gdf = GeoDataFrame(final_df, geometry = geometry)

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax = world.plot(figsize = (15,15)), marker = 'o', color = 'red', markersize = 15)
```

## Conclusion
Most restaurants of Zomato are available in India.
